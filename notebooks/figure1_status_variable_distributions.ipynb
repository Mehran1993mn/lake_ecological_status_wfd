## Figure caption

**Figure 1. Differences in lake water-quality and morphometric variables among ecological status classes.** Each panel shows the distribution of one variable across four ecological status groups: Bad/Poor, Moderate, Good, and High. Right-sided half-violins show the shape of each distribution, while the overlaid percentile boxes summarize the 25th to 75th percentiles, median, 5th to 95th percentile range, and mean. Kruskal-Wallis tests summarize whether the variable differs among status classes, and pairwise Mann-Whitney U tests are shown when the overall test is significant. Effect size is reported as absolute delta, where larger values indicate stronger separation between two status classes.
